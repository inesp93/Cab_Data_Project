```python
city_data=pd.read_csv(r'C:\Users\HT-ICT\Desktop\New\City.csv')
City = pd.DataFrame(city_data, columns= ['City','Population', 'Users'])
print(City)
City.isnull().sum() #no null values
City = City.drop_duplicates(subset=['City']) #removing duplicates based on city
City['Population'].dtypes
```

```python
def age(x):
    if x < 30:
        return 'younger than 30'
    elif x > 60:
        return 'older than 60'
    else:
        return 'between 30 and 60'
    
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
import numpy as np

df = df[['Company', 'year', 'month', 'Income_class', 'Age', 'Payment_Mode', 'City', 'Price Charged', 'Cost of Trip', 'KM Travelled', 'Profit']]
train = df.loc[:, df.columns != 'Profit']
label = df[['Profit']]
features = pd.get_dummies(train)

labels = np.array(label)
feature_list = list(features.columns)
features=np.array(features)
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.10, random_state = 42)

lr = LinearRegression()
lr.fit(train_features, train_labels);
predictions = lr.predict(test_features)
```
